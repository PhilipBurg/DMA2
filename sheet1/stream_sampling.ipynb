```python
# a) Reservoir Sampling:
data_stream = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
r = [0.64, 0.42, 0.58, 0.83, 0.43, #r[0]-r[4] as decisions for this exercise
     0.66, 0.66, 0.72, 0.7, 0.36,  #r[5]-r[9] as positions 
     0.56, 0.96, 0.74, 0.99, 0.85,
     0.91, 0.87, 0.88, 0.23, 0.32]
k = 5
S = []

# Initialize first k elements in the sample
for i in range(k):
    S.append(data_stream[i])
    print(f"Select o{i+1} with probability 5/{i+1} > 1")
    print(f"S(t={i+1}) = {S}")

# now sampling with alternating r values
r_index = 0  # Use pairs of r[i], r[i+1]

for i in range(k, len(data_stream)):
    t = i + 1
    prob = k / t
    r_decision = r[r_index]
    r_pos = r[r_index + 1]

    if r_decision < prob:
        pos = int(r_pos * k)
        print(f"Select o{t} with probability 5/{t} = {prob:.3f} > {r_decision:.2f} and add at position ⌊{r_pos:.2f} · {k}⌋ = {pos}")
        S[pos] = data_stream[i]
    else:
        print(f"Select o{t} with probability 5/{t} = {prob:.3f} < {r_decision:.2f} and skip")

    print(f"St={t} = {S}")
    r_index += 2  # move to next pair
```

```python
#now min-wise sampling from b)
sample = []

for t in range(10):
    sample.append((r[t], data_stream[t]))
    smallest_k = sorted(sample)[:k]
    result = [item for _, item in smallest_k]
    print(f"S(t={t+1}) = {result}")
```
